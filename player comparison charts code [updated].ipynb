```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
```

```python
# Data for players
data = {
    'Metric': ['Pass Accuracy %', 'Long Ball Accuracy %', 'Dispossessed', 'Tackles Won %', 
               'Interceptions', 'Duels Won %', 'Aerial Duels Won %', 'Recoveries', 'Dribbled Past'],
    'M. Zubimendi': [75.5, 47, 84.1, 28.5, 75.5, 72.8, 89.4, 52.3, 85.4],
    'A. Mac Allister': [75.2, 64.3, 33.3, 38, 71.3, 57.4, 72.9, 82.9, 6.2],
    'C. Jones': [93.8, 98.2, 3, 18.6, 7.1, 31.9, 18.6, 15.9, 49.6]
}

# Creating DataFrame
df = pd.DataFrame(data)
```

```python
# Number of variables
categories = list(df['Metric'])
N = len(categories)

# What will be the angle of each axis in the plot? (we divide the plot / number of variables)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
```

```python
# Number of variables
categories = list(df['Metric'])
N = len(categories)

# What will be the angle of each axis in the plot? (we divide the plot / number of variables)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


```

```python
# Titles and other customization details
T = dict(
    title_name='M. Zubimendi',
    title_color='blue',
    subtitle_name='Player 1',
    subtitle_color='blue',
    title_name_2='A. Mac Allister',
    title_color_2='orange',
    subtitle_name_2='Player 2',
    subtitle_color_2='orange',
    title_name_3='C. Jones',
    title_color_3='green',
    subtitle_name_3='Player 3',
    subtitle_color_3='green'
)

note = 'Visual by - @mizgans \nFotmob Percentile data per 90 numbers \n The Tomkins Times'

# Function to create radar chart with specified adjustments and white background
def create_radar_chart_white(df, players, T, note):
    plt.figure(figsize=(10, 10))
    ax = plt.subplot(111, polar=True)
    
    # Define colors and alphas
    colors = [T['title_color'], T['title_color_2'], T['title_color_3']]
    alphas = [0.76, 0.6, 0.5]
    
    for idx, player in enumerate(players):
        values = df[player].values.flatten().tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=player, color=colors[idx], alpha=alphas[idx])
        ax.fill(angles, values, color=colors[idx], alpha=0.25)
    
    plt.xticks(angles[:-1], categories, color='black', size=12)
    ax.set_rlabel_position(30)
    plt.yticks([20, 40, 60, 80, 100], ["20", "40", "60", "80", "100"], color="black", size=8)
    plt.ylim(0, 100)
    
    # Adding title
    plt.title(f"{T['title_name']} vs {T['title_name_2']} vs {T['title_name_3']} - Percentile Comparison 2023/24 Season", color='#654321', size=15, pad=40, fontweight='bold', fontname='Arial')
    
    # Adding note at the bottom right
    plt.text(1.05, -0.05, note, horizontalalignment='center', color='brown', size=11, transform=ax.transAxes)
    
    # Adding legends with specified colors at the top right
    legend = plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=11)
    for text, color in zip(legend.get_texts(), colors):
        text.set_color(color)
    
    plt.show()

# Plotting the radar chart with adjustments and white background
create_radar_chart_white(df, ['M. Zubimendi', 'A. Mac Allister', 'C. Jones'], T, note)
```
